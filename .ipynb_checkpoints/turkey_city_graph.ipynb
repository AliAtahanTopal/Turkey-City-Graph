# Algorithms

## Breadth First Search

```python
def BFS(initial, goal, graph):
  frontier = [initial]
  explored = []
  pred = {}
  found = False
  while len(frontier) > 0:
    state = frontier.pop(0)
    explored.append(state)
    if goal == state:
      found = True
      break
    for neighbor in graph[state]:
      if (neighbor not in frontier) & (neighbor not in explored):
        frontier.append(neighbor)
        pred[neighbor] = state
  if found == True:
    output = 'SUCCESS'
    start = goal
    path = [start]
    while start != initial:
      prev = pred[start]
      path.append(prev)
      start = prev
    path.reverse()
  else:
    output = 'FAILURE'
  return output, path, explored
```

## Depth First Search

```python
def DFS(initial, goal, graph):
  frontier = [initial]
  explored = []
  pred = {}
  found = False
  while len(frontier) > 0:
    state = frontier.pop()
    explored.append(state)
    if goal == state:
      found = True
      break
    for neighbor in graph[state]:
      if (neighbor not in frontier) & (neighbor not in explored):
        frontier.append(neighbor)
        pred[neighbor] = state
  if found == True:
    output = 'SUCCESS'
    start = goal
    path = [start]
    while start != initial:
      prev = pred[start]
      path.append(prev)
      start = prev
    path.reverse()
  else:
    output = 'FAILURE'
  return output, path, explored
```

## A*

```python
def astar (initial, goal, graph):
  h = distance_cities[goal][distance_cities['İL ADI'] == initial].values[0]
  frontier =[]
  heapq.heappush(frontier, (h, initial))
  explored = []
  pred = {}
  found = False
  cost = 0
  previous_city = (h, initial)
  while len(frontier) > 0:          
    state = heapq.heappop(frontier)
    cost += distance_cities[state[1]][distance_cities['İL ADI'] == previous_city[1]].values[0]
    previous_city = state 
    explored.append(state[1])
    if goal == state[1]:
      found = True
      break    
    for neighbor in graph[state[1]]:
      if (neighbor not in frontier) & (neighbor not in explored):        
        h1 = distance_cities[goal][distance_cities['İL ADI'] == neighbor].values[0]
        frontier.append(((h1 + cost), neighbor))
        heapq.heapify(frontier)
        pred[neighbor] = state[1]       
  if found == True:
    output = 'SUCCESS'
    start = goal
    path = [start]
    while start != initial:
      prev = pred[start]
      path.append(prev)
      start = prev
    path.reverse()
  else:
    output = 'FAILURE'
  return output, path, explored
```

## Comparison Method

```python
def tries (num_of_tries):
  all_bfs_visited = []
  all_dfs_visited = []
  all_astar_visited = []
  all_path_bfs = []
  all_path_dfs = []
  all_path_astar =[]
  path_distance_bfs = []
  path_distance_dfs = []
  path_distance_astar = []

  for i in range(num_of_tries):
    rand1 = random.randint(0,(len(distance_cities) - 1))
    rand2 = random.randint(0,(len(distance_cities) - 1))
    city1 = distance_cities['İL ADI'][rand1]
    city2 = distance_cities['İL ADI'][rand2]

    output_bfs, path_bfs, explored_bfs = BFS(city1, city2, graph)
    output_dfs, path_dfs, explored_dfs = DFS(city1, city2, graph)
    output_astar, path_astar, explored_astar = astar(city1, city2, graph)
    distance_bfs = 0
    for i in range(1, len(path_bfs)):
      tmp  = distance_cities[path_bfs[i-1]][distance_cities['İL ADI'] == path_bfs[i]].values
      distance_bfs += tmp
    
    distance_dfs = 0
    for i in range(1, len(path_dfs)):
      tmp  = distance_cities[path_dfs[i-1]][distance_cities['İL ADI'] == path_dfs[i]].values
      distance_dfs += tmp
    
    distance_astar = 0
    for i in range(1, len(path_astar)):
      tmp  = distance_cities[path_astar[i-1]][distance_cities['İL ADI'] == path_astar[i]].values
      distance_astar += tmp
    
    path_distance_bfs.append(distance_bfs)
    path_distance_dfs.append(distance_dfs)
    path_distance_astar.append(distance_astar)
    all_bfs_visited.append(len(explored_bfs))
    all_dfs_visited.append(len(explored_dfs))
    all_astar_visited.append(len(explored_astar))
    all_path_astar.append(len(path_astar))
    all_path_bfs.append(len(path_bfs))
    all_path_dfs.append(len(path_dfs))

  avg_bfs_path = np.mean(all_path_bfs)
  avg_dfs_path = np.mean(all_path_dfs)
  avg_astar_path = np.mean(all_path_astar)
  avg_dfs_visited = np.mean(all_dfs_visited)
  avg_bfs_visited = np.mean(all_bfs_visited)
  avg_astar_visited = np.mean(all_astar_visited)
  avg_dfs_distance = np.mean(path_distance_dfs)
  avg_bfs_distance = np.mean(path_distance_bfs)
  avg_astar_distance = np.mean(path_distance_astar)
  df = pd.DataFrame({'Wanted\Algorithm': ['Average Nodes Visited Length', 'Average Solution Path Length', 'Average Solution Distance in KM'],
                    'BFS': [avg_bfs_visited, avg_bfs_path, avg_bfs_distance],
                    'DFS': [avg_dfs_visited, avg_dfs_path, avg_dfs_distance],
                    'A*': [avg_astar_visited, avg_astar_path, avg_astar_distance],})
  return df
```
